## Context

Diabetes is one of the most frequent diseases worldwide and the number of diabetic patients are growing over the years. The main cause of diabetes remains unknown, yet scientists believe that both genetic factors and environmental lifestyle play a major role in diabetes.

A few years ago research was done on a tribe in America which is called the Pima tribe. In this tribe, it was found that the ladies are prone to diabetes very early. Several constraints were placed on the selection of these instances from a larger database. In particular, all patients were females at least 21 years old of Pima tribe.

## Data Dictionary
-------------------------

The dataset has the following information:

* Pregnancies: Number of times pregnant
* Glucose: Plasma glucose concentration over 2 hours in an oral glucose tolerance test
* BloodPressure: Diastolic blood pressure (mm Hg)
* SkinThickness: Triceps skin fold thickness (mm)
* Insulin: 2-Hour serum insulin (mu U/ml)
* BMI: Body mass index (weight in kg/(height in m)^2)
* DiabetesPedigreeFunction: A function that scores the likelihood of diabetes based on family history.
* Age: Age in years
* Outcome: Class variable (0: a person is not diabetic or 1: a person is diabetic)

## Q9. What do the summary statistics of the data represent? Find the summary statistics for all variables except 'Outcome' in the `pima` data. Take one column/variable from the output table and explain all its statistical measures. (3 Marks)

```python
pima.iloc[:,0:8].describe()
```

### Ans 9:
__count:__ The total number of entries in that column. Ex. There are 768 data points regarding the glucose level in the patients.

__mean:__ The average value in that column type, calculated as $\frac{1}{n}\sum_{i=1}^{n}x_i$ where n=768 is the number of entries and $x_i$ is the value of entry $i$. Ex. If we added up the total glucose level for all patients and divide by the number of patients in this dataset, we find an average glucose level of approximately 121.68. 

__std = Standard Deviation, $\sigma$:__ A statistic describing how dispersed the dataset is. If we take $\mu$ as the calculated mean and use the variables as described above, then $\sigma=\sqrt{\frac{\sum_{i=0}^n ( x_i - \mu )}{n}}$. In a Gaussian distribution, approximately 68% of the data fall within the range $\mu-\sigma$ to $\mu+\sigma$. Ex. With regard to glucose level in this dataset, $\sigma\approx30.44$.

__min = minimum value:__ This is the smallest value in the dataset for the given variable. Ex. There is at least one patient who had a glucose level of 44 and no patients with a glucose level less than 44. 

__25% = First Quartile:__ This is the threshold at which 25% of the data have smaller values. Ex. 25% of the patients (192) have a glucose level below 99.75.

__50% = Second Quartile = Median:__ This is the threshold at which half of the data have smaller values and half of the data have larger values. Ex. 50% of the patients (384) have a glucose level less than 117.00.

__75% = Third Quartile:__ This is the threshold at which three quarters of the data have smaller values. Ex. 75% of the patients (576) have glucose levels below 140.25.

__max = maximum value:__ This is the largest value in the dataset for the gien variable. Ex. There is at least one patient who had a glucose level of 199 and there are no patients who had a glucose level higher than 199.

## Q 10. Plot the distribution plot for the variable 'BloodPressure'. Write detailed observations from the plot. (2 Marks)

```python
sns.displot(pima['BloodPressure'], kind='kde')

import scipy.stats as stats
varname='BloodPressure'
BPmin  = pima[varname].min()
BPmax  = pima[varname].max()
BPmean = pima[varname].mean()
BPstd  = pima[varname].std()
x = np.linspace(BPmin,BPmax,300)
plt.plot(x, stats.norm.pdf(x, BPmean, BPstd),color='red')
plt.show()
```

### Ans 10:
In the Pima dataset, the BloodPressure variable roughly traces a Gaussian distribution, though the data are more sharply peaked and skewed to lower values. On the high side there is a small secondary peak, or shoulder, of patients with a blood pressure between 100 and 115.

## Q15. Create a pairplot for the variables 'Glucose', 'SkinThickness', and 'DiabetesPedigreeFunction'. Write your observations from the plot. (4 Marks)

```python
sns.pairplot(data=pima,vars=['Glucose', 'SkinThickness', 'DiabetesPedigreeFunction'], markers=['*','h'], hue='Outcome')
plt.show()
```

```python
data0 = pima[pima['Outcome']==0]
data1 = pima[pima['Outcome']!=0]
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,5))
ax1.xaxis.grid(color='gray', linestyle='dashed')
ax2.xaxis.grid(color='gray', linestyle='dashed')
ax3.xaxis.grid(color='gray', linestyle='dashed')

sns.kdeplot(data0['Glucose'], ax=ax1, label="Outcome = 0")
sns.kdeplot(data1['Glucose'], ax=ax1, label="Outcome = 1")
ax1.legend()

sns.kdeplot(data0['SkinThickness'], ax=ax2, label="Outcome = 0")
sns.kdeplot(data1['SkinThickness'], ax=ax2, label="Outcome = 1")
ax2.legend()

sns.kdeplot(data0['DiabetesPedigreeFunction'], ax=ax3, label="Outcome = 0")
sns.kdeplot(data1['DiabetesPedigreeFunction'], ax=ax3, label="Outcome = 1")
ax3.legend()
```

### Ans 15:
#### Single Variable Plots:

__Glucose:__ Patients with higher levels of glucose are more strongly associated with a positive outcome than patients lower levels. The glucose distribution for patients with a positive outcome appears to be made from two overlapping Gaussians with peaks at around 140 and 180 while the distribution for negative outcomes is tracks well with a Gaussian centered around 180. In the negative distribution, a small secondary peak is visible around 180 as well, though it is much smaller than the 180 peak in the positive distribution. It may be interesting to investigate the origin of the peak at 180 to see if there is a reason why it is so much more pronounced in the positive patients than the negative ones. 

__Skin Thickness:__ The distribution of skin thickness for patients with positive and negative outcomes are similar in shape, and both appear to be made from multiple underlying Gaussians. For both positive and negative patients, the main peak is around a skin thickness of 20, and in both cases the drop off on the high side is interrupted by a secondary distribution. 

__Diabetes Pedigree Function (DPF):__ For both positive and negative patients, the DPF is peaked around 0.25, though the positive distribution is shifted slightly to larger values. There is a steep dropoff to the small side for both the positive and negative distributions, and on the high side both distributions fall slower than would be expected for a Gaussian. Proportionally more mositive patients stay with higher values of DPF than do negative patients.

#### Scatter Plots:

__Glucose v Skin Thickness:__ The histograms for skin thickness in positive and negative patients are very similar, and adding glucose to make a scatter plot does little to obviously differentiate the two patient types. As would be expected from the glucose histogram, patients with higher glucose levels are more likely to be positive and patients with lower glucose levels are more likely to be negative, but there is no clear relationship between glucose level and skin thickness. There is one very notable outlier which has a high glucose level and high skin thickness and this person tested as positive. 

__Glucose v DPF:__ The trend of higher glucose values corresponding to higher probabilities of positivity is also visible in this scatter distribution, but there is again little obvious correlation between glucose level and DPF for positive or negative patients. There are a handful of positive outliers with relatively high glucose levals and high DPF but there is also a negative outlier with high glucose and high DPF. 

__Skin Thickness v DPF:__ Skin thickness and DPF both have very similar positive patient and negative patient shapes in their individual histograms, so it is unsurprising that making a scatter plot of the two variables has substantial overlap between positive and negative patients. While most of the outliers are positive patients with relatively thick skin and a very high DPF, it is also notable that the patient with the thickest skin was also positive and yet had a DPF close to the mean.
